```python
import cv2
import matplotlib.pyplot as plt

# Correct image path (IMPORTANT)
image_path = '/content/Screenshot 2025-12-15 191303.png'

# Read image
img = cv2.imread(image_path)

# Safety check
if img is None:
    raise ValueError("Image not loaded! Check filename or upload step.")

# Convert BGR to RGB for display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# Detect faces
faces = face_cascade.detectMultiScale(
    gray,
    scaleFactor=1.1,
    minNeighbors=5
)

# Draw rectangles
for (x, y, w, h) in faces:
    cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)

# Display result
plt.figure(figsize=(8,6))
plt.imshow(img_rgb)
plt.axis('off')
plt.title(f'Faces Detected: {len(faces)}')
plt.show()

print("Number of faces detected:", len(faces))
```

```python
import cv2
import matplotlib.pyplot as plt
import os

# Load Haar Cascade
face_cascade = cv2.CascadeClassifier(
    cv2.data.haarcascades + 'haarcascade_frontalface_default.xml'
)

# Get all files in /content
files = os.listdir('/content')

# Allowed image extensions
image_extensions = ('.jpg', '.jpeg', '.png')

for file in files:
    if file.lower().endswith(image_extensions):
        print(f"\nProcessing image: {file}")

        image_path = os.path.join('/content', file)
        img = cv2.imread(image_path)

        if img is None:
            print(" Could not read image")
            continue

        # Convert for processing & display
        img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

        # Detect faces
        faces = face_cascade.detectMultiScale(
            gray,
            scaleFactor=1.1,
            minNeighbors=5
        )

        # Draw rectangles
        for (x, y, w, h) in faces:
            cv2.rectangle(img_rgb, (x, y), (x+w, y+h), (0, 255, 0), 2)

        # Display result
        plt.figure(figsize=(6,5))
        plt.imshow(img_rgb)
        plt.axis('off')
        plt.title(f"{file} | Faces Detected: {len(faces)}")
        plt.show()

        print(" Number of faces detected:", len(faces))
```
